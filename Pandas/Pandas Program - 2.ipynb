#### 25)	 Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe)

```python
def youngest_reachHR(df,hr):
    temp = df.groupby(['playerID']).sum()
    temp = temp[temp['HR'] >= hr]
    temp = temp.reset_index(['playerID'])
    players = temp['playerID']
    
    temp = df.groupby(['playerID','nameFirst','nameLast','birthYear','yearID']).sum()
    temp = temp.reset_index(['playerID','nameFirst','nameLast','birthYear','yearID'])
    
    players = list(players)
    players.sort()
    agelist = []
    for i in players:
        tp = temp[temp['playerID']==i]
        tp = tp.sort_values('yearID')
        HR = 0
        for j in range(tp.shape[0]):
            HR += list(tp['HR'])[j]
            if HR >= hr:
                year = list(tp['yearID'])[j]
                age = year - list(tp['birthYear'])[j]
                agelist.append(age)
                break
        if HR < hr:
            agelist.append(np.nan)
    
    min_age = min(agelist)
    playerReach =[]
    for i in range(len(agelist)):
        if agelist[i] == min_age:
            playerReach.append(i)
    playerID = []
    for i in playerReach:
        playerID.append(players[i])
    return playerID
```

```python
playerID = youngest_reachHR(df,300)
datas = []
for i in playerID:
    datas.append(df[df['playerID']==i])
temp = pd.concat(datas)
temp = temp.groupby(['playerID','nameFirst','nameLast']).sum()
temp
```
