```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.mixture import GaussianMixture
np.random.seed(3)
import matplotlib.pyplot as plt
%matplotlib inline  
from sklearn.cluster import KMeans
from datetime import datetime
from sklearn.decomposition import PCA
from sklearn.decomposition import FastICA
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection, johnson_lindenstrauss_min_dim
from sklearn import tree
import graphviz
from sklearn import svm
import seaborn as sn
import tensorflow as tf
tf.set_random_seed(3)
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import learning_curve
```

```python
data = pd.read_csv('../Data/OnlineNewsPopularity.csv')

# Data Normalization
X = data[data.columns[1:-1]]

X = X.values #returns a numpy array
min_max_scaler = MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X)
X = pd.DataFrame(x_scaled,columns=data.columns[1:-1])
y = data[' shares']
y[y <1400] = 0
y[y >=1400] = 1


X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30, random_state=2)
```

```python
init = ['k-means++','random']
algorithm = ['full','elkan']
precompute_distance = [True,False]
```

```python
k_init_score = []
for initial in init:
    for i in range(1,21):
        Kmean = KMeans(n_clusters=i,random_state=3,init=initial)
        Kmean.fit(X_train)
        k_init_score.append(Kmean.inertia_)
```

```python
plt.plot(k_init_score[:20],'r',label='K-means++')
plt.plot(k_init_score[-20:],'b',label='random')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different initialization method')
```

```python
k_init_score = []
k_test_score = []
for i in range(1,21):
    Kmean = KMeans(n_clusters=i,random_state=3)
    Kmean.fit(X_train)
    k_init_score.append(Kmean.inertia_)
    k_test_score.append(Kmean.score(X_test))
```

```python
plt.plot(k_init_score,'b')
# plt.plot(k_test_score,'b',label='random')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different K')
```

```python
k_alg_score = []
for alg in algorithm:
    for i in range(1,11):
        Kmean = KMeans(n_clusters=i,random_state=3,algorithm=alg)
        Kmean.fit(X_train)
        k_alg_score.append(Kmean.inertia_)
plt.plot(k_alg_score[:10],'r',label = 'full')
plt.plot(k_alg_score[-10:],'b',label = 'elkan')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different Alogrithm')
```

```python
k_pre_score = []
for pre_cal in precompute_distance:
    start = datetime.now()
    for i in range(1,11):
        Kmean = KMeans(n_clusters=i,random_state=3,precompute_distances=pre_cal)
        Kmean.fit(X_train)
        k_pre_score.append(Kmean.inertia_)
    print('Precompute_distance is:',str(pre_cal))
    print('Time:',datetime.now()-start)
plt.plot(k_pre_score[:10],'r',label='True')
plt.plot(k_pre_score[-10:],'b',label='False')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance Difference')
```

```python
k_tol_score = []
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,max_iter=30000,tol=1e-7)
    Kmean.fit(X_train)
    k_tol_score.append(Kmean.inertia_)
    
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,max_iter=30000,tol=100)
    Kmean.fit(X_train)
    k_tol_score.append(Kmean.inertia_)
    
plt.plot(k_tol_score[:10],'r',label='tol=1e-7')
plt.plot(k_tol_score[-10:],'b',label='tol=100')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different tolerance')
```

```python
k_max_iter_score = []
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,max_iter=1,tol=1e-7)
    Kmean.fit(X_train)
    k_max_iter_score.append(Kmean.inertia_)
    
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,max_iter=3000,tol=1e-7)
    Kmean.fit(X_train)
    k_max_iter_score.append(Kmean.inertia_)
    
plt.plot(k_max_iter_score[:10],'r',label='max_iter=1')
plt.plot(k_max_iter_score[-10:],'b',label='max_iter=3000')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different max of iteration')
```

```python
k_init_state_score = []
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,n_init=1)
    Kmean.fit(X_train)
    k_init_state_score.append(Kmean.inertia_)
    
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,random_state=3,n_init=30)
    Kmean.fit(X_train)
    k_init_state_score.append(Kmean.inertia_)
    
plt.plot(k_init_state_score[:10],'r',label='n_init=1')
plt.plot(k_init_state_score[-10:],'b',label='n_init=30')
plt.legend(loc="best")
plt.xlabel('K')
plt.ylabel('Sum of Squared Distances from Center')
plt.title('Performance in Different Number of Time with Different Centroid Seeds')
```

```python
GBDT_PCA = []
DT_PCA = []
for i in range(1,41):
    pca = PCA(n_components=i,random_state=3,whiten=True)
    pca.fit(X_train)
    pca_x_train = pca.transform(X_train)
    pca_x_test = pca.transform(X_test)
    if i < 10:
        max_ = i
    else:
        max_ = 10
    clf = GradientBoostingClassifier(random_state=3,max_depth=1,max_features=max_,learning_rate=0.1)
    clf.fit(pca_x_train,y_train)
    GBDT_PCA.append(clf.score(pca_x_train,y_train))
    
    clf_DT = tree.DecisionTreeClassifier(criterion='gini',random_state=3,max_depth=2,max_leaf_nodes=80)
    clf_DT.fit(pca_x_train,y_train)
    DT_PCA.append(clf.score(pca_x_test,y_test))
```

```python
plt.plot(GBDT_PCA,label='Grandient Boosted')
plt.plot(DT_PCA,label='Decision Tree')
plt.legend(loc="best")
plt.xlabel('Number of Components')
plt.ylabel('Accuracy')
plt.title('Performance in PCA')
```

```python
node = 50
n_node_hl1 = node

n_classes = 2
batch_size = 250

x_ph = tf.placeholder('float', [None,len(feature_list)])
y_ph = tf.placeholder('int32',[None,2])   ## int didn't work

def nn_model_selection(data):
    hidden_1 = {'weights': tf.Variable(tf.random_normal([23,n_node_hl1])),
               'biases': tf.Variable(tf.random_normal([n_node_hl1]))}

    output_l = {'weights': tf.Variable(tf.random_normal([n_node_hl1,n_classes])),
             'biases': tf.Variable(tf.random_normal([n_classes]))}


    l1 = tf.add(tf.matmul(data,hidden_1['weights']),hidden_1['biases'])
    l1 = tf.nn.sigmoid(l1)

    output = tf.add(tf.matmul(l1, output_l['weights']),output_l['biases'])

    return output
```

```python
accuracy_list=[]
train_list = []

def train_nn_selection(x_ph,model):
    pred = model(x_ph)
    cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits_v2(logits=pred, labels=y_ph))
    optimizer = tf.train.AdamOptimizer(learning_rate=0.005).minimize(cost)
    
    epoch = 100
    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        for ep in range(epoch):
            epoch_loss = 0
            for i in range(int(X_train.shape[0]/batch_size)):
                beg = i*batch_size
                end = beg + batch_size
                ep_x = X_train_nn[beg:end]
                ep_y = y_train_nn[beg:end]
                qq, c = sess.run([optimizer,cost],feed_dict= {x_ph:ep_x,y_ph:ep_y})
                epoch_loss += c

        correct = tf.equal(tf.argmax(pred,1),tf.argmax(y_ph,1)) 
        accuracy = tf.reduce_mean(tf.cast(correct,'float'))
        Acc = accuracy.eval({x_ph:X_train_nn, y_ph:y_train_nn})
        train_list.append(Acc)
        Acc = accuracy.eval({x_ph:X_test_nn, y_ph:y_test_nn})
        accuracy_list.append(Acc)
        print('Acc:',Acc)
```
